```python
import torch 
```

```python
random_tensor = torch.rand(5)
print(random_tensor)
print(random_tensor.ndim)
```

```python
random_tensor = torch.rand(2,5,3)
print(random_tensor)
print(random_tensor.ndim)
print(random_tensor.shape)
```

```python
torch.manual_seed(42)
random_tensor = torch.rand(5)
print(random_tensor)
```

```python
# Generate random integers 
# torch.randint(low, high, size)
random_integer = torch.randint(0,10,(5,3))
print(random_integer)
```

```python
# generate random numbers from a normal distribution 

# torch.normal(mean, std, size)

random_numbers_normal = torch.normal(0.5, 2, (5,))
print(random_numbers_normal)
```

```python
## Creating noisy image using pytorch

import matplotlib.pyplot as plt

# h,w of image 256x256

uniform_noise = torch.rand(256,256)

plt.imshow(uniform_noise, cmap='grey')
plt.title("Uniform Noise Image")
plt.axis('off')
plt.show()
```

```python
import matplotlib.pyplot as plt


uniform_noise_rgb = torch.rand(256,256,3)

plt.imshow(uniform_noise_rgb.numpy())
plt.title("Uniform Noise Image")
plt.axis('off')
plt.show()
```

```python
print("shape", uniform_noise_rgb.shape)
print("Data type", uniform_noise_rgb.dtype)
print("Min Value", torch.min(uniform_noise_rgb))
```
